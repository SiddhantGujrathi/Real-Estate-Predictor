## Problem 1
Now CHAS feature have 0 -> 471 and 1-> 35 , there are chances that train only contains the rows that have CHAS with 0 value.
-> then you have to use Stratified Sampling

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits = 1 , test_size=0.2, random_state=42)
for train_index , test_index in split.split(housing , housing['CHAS']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
```

# Missing Attributes

To take care of missing attributes , you have three options :
    <br>1. Get rid of the missing data points 
    <br>2. Get rid of the whole attribute
    <br>3. Set the value to 0, mean or median

```python
a = housing.dropna(subset=["RM"])  #Option1 , #Note that the original housing dataset will remain unchanged
a.shape
```

```python
housing.drop("RM",axis=1)  #Option2
#Note that there is no RM column and also housing is unchanged
```

```python
median = housing["RM"].median()    #Compute median for option3
```

```python
housing["RM"].fillna(median)     #Option 3
#Note that the original housing dataset will remain unchanged
```

```python
housing.shape
```

```python
housing.describe()    # Before we started filling missing attribute
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")

# Fitting the data to the imputer object
imputer.fit(housing)
```

```python
imputer.statistics_   # It founds median value for each column
```

```python
X = imputer.transform(housing)   #impute the missing values
```

```python
housing_tr = pd.DataFrame(X, columns=housing.columns)
```

```python
housing_tr.describe()
```

## Scikit-learn Design

Primarily , three types of objects:<br>
<br>1.Estimators - It estimates some parameter based on dataset eg.SimpleImputer
<br>It has fit method and transform method
<br>fit method-> fits the dataset and calculate internal parameters

<br>2.Transformers - transform method takes input and returns output based on the learnings from fit().
<br>It also has a convenience function called fit_transform() which fits and then transforms.

<br>3.Predictors - linear_regression model is an example of predictor.fit() and predict() are two common functions.
It will also gives score functions which evaluates the predictions.


## Feature Scaling

Primarily, two types of feature scaling methods:
1. Min-Max Scaling (Normalization)
    (value-min) / (max-min)
    Sklearn provides class called MinMaxScaler for this

2. Standardization
    (value-min)/std
    -->result variance is 1
    Sklearn provides class called StandardScaler for this

## Creating a Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

my_pipeline = Pipeline([
('imputer',SimpleImputer(strategy="median")),
('std_scaler',StandardScaler()),
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing)  # it fits and transforms till last.
```

```python
housing_num_tr  # it is an numpy array
```

```python
housing_num_tr.shape
```

## Selecting a desired model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(housing_num_tr,housing_labels)
```

```python
some_data = housing.iloc[:5]
```

```python
some_labels = housing_labels.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

```python
model.predict(prepared_data)
```

```python
list(some_labels)
```

## Evaluating data

```python
import numpy as np
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels,housing_predictions)
rmse = np.sqrt(mse)
```

```python
# for lin_mse LinearRegression model = 23.654845182413293
rmse
```

it is overfitting , if rmse =0.0

## Using Better Evaluation Technique - Cross Validation

```python
#kFold
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model , housing_num_tr,housing_labels,scoring="neg_mean_squared_error",cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(rmse_scores):
    print("Scores:",scores)
    print("Mean: ", scores.mean)
    print("Standard Deviation : ", scores.std)
    
    
```

```python
print_scores(scores)
```

## Testing the model on test data

```python
X_test = strat_test_set.drop("MEDV",axis=1)
Y_test = strat_test_set["MEDV"].copy()

X_test_prepared = my_pipeline.transform(X_test)

final_predictions = model.predict(X_test_prepared)
final_rmse = mean_squared_error(Y_test,final_predictions)
final_rmse = np.sqrt(final_rmse)
final_rmse
```
